```python
# Buat script hello_tff.py
with open("hello_tff.py", "w") as f:
    f.write("""
import tensorflow_federated as tff

# Cek versi TFF
print('TFF version:', tff.__version__)

# Buat federated computation sederhana
hello = tff.federated_computation(lambda: 'Hello World')

# Jalankan computation
print(hello())
""")
```
